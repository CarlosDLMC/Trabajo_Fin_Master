```python
from datetime import datetime, timedelta

def date_generator(start, end=datetime.now()):
    month = timedelta(days=150)
    date_dt = datetime.strptime(start, "%Y-%m-%d")
    while date_dt < end:
        previous = date_dt
        date_dt += month
        date_dt = min(date_dt, end)
        yield previous.strftime("%Y-%m-%d"), date_dt.strftime("%Y-%m-%d")
```

```python
from Historic_Crypto import HistoricalData

sym_date = {
    "BTC": "2015-07-20",
    "ETH": "2016-05-18",
    "ADA": "2021-03-18",
    "LINK": "2019-06-27",
    "DOGE": "2021-06-03",
    "LTC": "2016-08-17",
    "DOT": "2021-06-16",
    "SOL": "2021-09-15",
    "ALGO": "2019-08-15",
}

simbolo_moneda = {
    "BTC": "Bitcoin",
    "ADA": "Cardano",
    "LINK": "Chainlink",
    "DOGE": "Dogecoin",
    "ETH": "Ethereum",
    "LTC": "Litecoin",
    "DOT": "Polkadot",
    "SOL": "Solana",
    "ALGO": "Algorand"
}

counter = 0
sym = "BTC"
for date in date_generator(sym_date[sym]):
    counter += 1
    print(f"\n\n\nDF number: {counter}. From {date[0]} to {date[1]}\n\n\n")
    data = HistoricalData(f'{sym}-USD', 300, start_date=f'{date[0]}-00-01', end_date=f'{date[1]}-00-00').retrieve_data()
    data.to_csv(f'bitcoin_5_mins/{simbolo_moneda[sym]}_{counter}.csv')
```
